```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Assuming df is already loaded and contains 3 columns
x_vals = df.iloc[:, 0].values  # Days
y_vals = df.iloc[:, 1].values  # Compressive strength
z_vals = df.iloc[:, 2].values  # Stiffness

# Create meshgrid for x and y
x, y = np.meshgrid(x_vals, y_vals)

# Reshape z_vals to broadcast with y
z_matrix = np.cos(np.sqrt(y**2 + z_vals[np.newaxis, :]**2)) * np.exp(-0.1*(x**2 + y**2))

# Plotting
fig = plt.figure(figsize=(17, 11))
ax = fig.add_subplot(111, projection='3d')

ax.view_init(elev=5, azim=100)

ax.text(10, 25, 0.5, 'Point of Interest', color='green')

surf = ax.plot_surface(x, y, z_matrix, cmap='coolwarm', edgecolor='none')

fig.colorbar(surf, shrink=0.5, aspect=10)

ax.set_xlabel('(X-axis) Days')
ax.set_ylabel('(Y-axis) Compressive Strength (MPa)')
ax.set_zlabel('(Z-axis) Stiffness k (Kns/m)')
ax.set_title('3D Surface Plot')

plt.show()
```

```python
import plotly.graph_objects as go
import numpy as np
import pandas as pd

# Assuming df is already defined
x = df.iloc[:, 0].values  # Days
y = df.iloc[:, 1].values  # Compressive strength
z = df.iloc[:, 2].values  # Stiffness

# Create meshgrid (required for surface)
x_grid, y_grid = np.meshgrid(x, y)

# Create surface Z values using your equation (synthetic)
z_surface = np.cos(np.sqrt(y_grid**2 + z[np.newaxis, :]**2)) * np.exp(-0.1*(x_grid**2 + y_grid**2))

# Create Plotly surface
fig = go.Figure(data=[
    go.Surface(z=z_surface, x=x_grid, y=y_grid, colorscale='Plasma')
])

# Customize the camera (like eye = x:2, y:2, z:1.5)
fig.update_layout(
    scene=dict(
        xaxis_title='Days',
        yaxis_title='Compressive Strength (MPa)',
        zaxis_title='Stiffness (kN/s/m)',
        camera=dict(eye=dict(x=2, y=2, z=1.5))  # Control 3D viewing angle
    ),
    title='3D Surface Plot with Custom Camera View'
)

fig.show()
```
